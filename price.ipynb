```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
d_mu, d_sigma = 30, 10
p_mu = np.random.uniform(low=20,high=100)

def sigmoid(x):
    return 1 / (1 + np.exp(-x/500))

def demand(t):
    return np.random.normal(d_mu, d_sigma)

def price(t):
    p_low = 20 + np.abs(np.sin(t)) * np.random.random() * 180
    p_high = 300 - np.abs(np.sin(t)) * np.random.random() * 100
    if np.random.random() >= 0.6:
        p_final = 20 + np.random.random() * 10
    else:
        p_final = np.random.uniform(low=p_low,high=p_high)
    return p_final
```

```python
T=200
prices=[0]
z=[]
for i in range(1,T+1):
    prices.append(price(i))
    z.append({"t": i, "price": prices[i], "price_avg": np.mean(prices)})
df=pd.DataFrame(z)
sns.lineplot(x="t",y="price",data=df)
```
